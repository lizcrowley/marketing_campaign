```python
# Import all packages
import pandas as pd # To process data
import matplotlib.pyplot as plt # To visualize data
import numpy as np # To perform calculations

# Read in the csv and store in a DataFrame
url = 'https://github.com/lizcrowley/marketing_campaign/blob/main/marketing_campaign.csv?raw=true'
market_data = pd.read_csv(url, delimiter = '\\t', 
                          usecols = [i for i in range(15)]) # Include relevant columns
market_data
```

```python
# Learn more about the data you're working with
market_data.dtypes
```

```python
# Use descriptive statistics to get an idea of what the data contains
market_data.describe(include = np.number)
```

```python
# Are there any null values?
market_data.isnull().sum()

# Replace any null values with '0'
market_data.replace(np.nan,'0',inplace = True)
```
